## Education level vs earnings
Checks whether earnings differ according to education level


```python
# importing data set
import pandas as pd
from scipy import stats
excel_file = 'PSID.csv'
earnings = pd.read_csv(excel_file)
```

## Creating two education categories from given categories
Two categories were created using given education categories as follows.
If given education category is greater than 12, new category is 1 and else new education category is 0

```python
# create response variable
earnings['educatn_category'] = ([1 if i> 12 else 0 for i in earnings.educatn]) # 1 if more than 12 in education, else 0
```

## Hypothesis
H0: There is no difference in average earnings between (created 2 categories mentioned above) two education categories.                          
Ha: There is a difference in average earnings between (created 2 categories mentioned above) two education categories.

```python
mean_0 = earnings.earnings [earnings.educatn_category == 0].mean()
mean_1 = earnings.earnings [earnings.educatn_category == 1].mean()
print(mean_0, mean_1)
mean_diff = mean_1-mean_0
print(mean_diff)
# get sample earnings sd
sd = earnings.earnings.std()
print(sd)

# compute t statistic
t = (mean_diff-0)/(sd/4856**(1/2))
print(t)

#compute p value
p = 2*(1 - stats.t.cdf(t,1))
print(p)
```

calculated p value = 0.0147 < 0.05                                 
Therefore we reject H0 in favor of Ha at 5% significance level.                       
Thus, the test suggests that there is a difference in average earnings between (created 2 eduation categories mentioned above) two education categories

## Other analysis

## Creating two earnings categories from given earnings

Two categories were created using given earnings as follows. If given earning is greater than 12000, new earnings category is 1 and else new earnings category is 0 .

```python
earnings['earnings_category'] = ([1 if i> 12000 else 0 for i in earnings.earnings]) # 1 if more than 12000 in earnings, else 0
```

```python
earnings.earnings_category.value_counts()
```

Crosstab was created as follows to check the observed frequencies (To check whether chi-squared test can be applied.)  

Hypothesis                                                          
H0: There is no relationship between eranings (created 2 categories mentioned above) and education categories.                  
Ha: There is a relationship between eranings (created 2 categories mentioned above) and education categories.

```python
contingency = pd.crosstab(earnings['earnings_category'], earnings['educatn_category'])
print(contingency)
```

All the observed frequencies are greater than 5. Therefore chi-squared test can be applied.

```python
stats.chi2_contingency(contingency)
```

calculated p value = 6.7526991470723395e-62 < 0.05                                
Therefore we reject H0 in favor of Ha at 5% significance level.                       
Thus, the test suggests that there is a relationship between (created categories above) earnings and education categories.
