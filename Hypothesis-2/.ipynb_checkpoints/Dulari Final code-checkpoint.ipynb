## Education level vs earnings
Checks whether earnings differ according to education level


```python
# importing data set
import pandas as pd
from scipy import stats
excel_file = 'PSID.csv'
earnings = pd.read_csv(excel_file)
```

## Creating two education categories from given categories
Two categories were created using given education categories as follows.
If given education category is greater than 12, new category is 1 and else new education category is 0

```python
# create response variable
earnings['educatn_category'] = ([1 if i> 12 else 0 for i in earnings.educatn]) # 1 if more than 12 in education, else 0
```

## Hypothesis
H0: There is no difference in average earanings between two education categories.                          
Ha: There is a difference in average earanings between two education categories.

```python
mean_0 = earnings.earnings [earnings.educatn_category == 0].mean()
mean_1 = earnings.earnings [earnings.educatn_category == 1].mean()
print(mean_0, mean_1)
mean_diff = mean_1-mean_0
print(mean_diff)
# get sample earnings sd
sd = earnings.earnings.std()
print(sd)

# compute t statistic
t = (mean_diff-0)/(sd/4856**(1/2))
print(t)

#compute p value
p = 2*(1 - stats.t.cdf(t,1))
print(p)
```
