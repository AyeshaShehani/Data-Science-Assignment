# AGE vs WORKING HOURS




## Checks whether age affects to the working hours

```python
# data import
import pandas as pd
from scipy import stats
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

file = 'dat.csv'
data = pd.read_csv(file)
```

# Categoarization of age

age>40 --> 1(ADULT)

other wise -->0 (YOUTH)


```python
# create response variable
data['age_cat'] = ([1 if i> 40 else 0 for i in data.age]) # 1 if more than 40 age, else 0
```

# Hypothesis

#### H0: There is no difference in average hours between  two age categories.(created 2 categories mentioned above)

#### Ha: There is a difference in average hours between  two age categories. (created 2 categories mentioned above) 

```python
mean_0 = data.hours [data.age_cat == 0].mean()
mean_1 = data.hours [data.age_cat == 1].mean()
print(mean_0, mean_1)
mean_diff = mean_1-mean_0
print(mean_diff)
# get sample earnings sd
sd = data.hours.std()
print(sd)

# compute t statistic
t = (mean_diff-0)/(sd/4856**(1/2))
print(t)

#compute p value
p = 2*(1 - stats.t.cdf(t,1))
print(p)
```

### calculated p value = 0.1158 > 0.05

##### Therefore we fail to  reject H0 in favor of Ha at 5% significance level.
#### Thus, the test suggests that there may be no difference in average hours between two age categories. (may be age doesnot affect the working hours )
