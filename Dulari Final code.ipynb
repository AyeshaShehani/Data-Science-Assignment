## Education level vs earnings
Checks whether earnings differ according to education level


```python
# importing data set
import pandas as pd
excel_file = 'C:\\Users\\199314V\\Data Science Assignment\\PSID.csv'
earnings = pd.read_csv(excel_file)
```

## Creating two education categories from given categories
Two categories were created using given education categories as follows.
If given education category is greater than 12, new category is 1 and else new education category is 0

```python
# create response variable
earnings['educatn_category'] = ([1 if i> 12 else 0 for i in earnings.educatn]) # 1 if more than 12 in education, else 0
```
